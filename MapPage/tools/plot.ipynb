```python
import json
import matplotlib.pyplot as plt
from shapely.geometry import Polygon
```

```python
# load data
with open("province_coordinates/12นนทบุรี.json", "r", encoding="utf-8") as f:
    data = json.load(f)
```

```python
data
```

```python
center_lat = float(data['center']['lat'])
center_lng = float(data['center']['lng'])
print(f"center ({center_lat} , {center_lng})")
```

```python

#get positions
coordinates = data['coordinates']
coordinates
```

```python
plot_coordinate = [] #A list of tuple
for blob in coordinates:
    for point in blob:
        point_tuple = (point['lat'], point['lng'])
        #print(point_tuple)  #debug
        plot_coordinate.append(point_tuple)
```

```python
plot_coordinate
```

```python
# polygon
polygon = Polygon(plot_coordinate)
# plotting
fig, ax = plt.subplots(figsize=(8, 8))
x, y = polygon.exterior.xy

ax.plot(y, x, color='blue', linewidth=2, label='Boundary')
ax.fill(y,x)
ax.scatter(center_lng,center_lat,  color='red', marker='x', label='Center Point')
ax.set_title(f"{data['code']}, resolution : {len(plot_coordinate)} points")
ax.set_xlabel("Longitude")
ax.set_ylabel("Longitude")
ax.set_xlabel("Latitude")
ax.legend()
ax.grid(True)

plt.axis("equal")
plt.show()
```

```python
# polygon
polygon = Polygon(plot_coordinate)
# plotting
fig, ax = plt.subplots(figsize=(8, 8))
ax.grid(True)

x, y = polygon.exterior.xy

ax.plot(y, x, color='blue', linewidth=2,)
ax.fill(y,x)
ax.scatter(center_lng,center_lat,  color='red', marker='x',)
ax.grid(False)
ax.set_title(f"")
ax.set_xlabel("")
ax.set_ylabel("")
ax.set_xlabel("")
ax.legend()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.get_xaxis().set_ticks([])
ax.get_yaxis().set_ticks([])

plt.axis("equal")
plt.savefig("12.jpeg", transparent=True)
plt.show()
```
